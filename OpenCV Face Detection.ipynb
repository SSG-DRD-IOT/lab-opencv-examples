# OpenCV Face Detection Tutorial

### What are Cascade Classifiers?

Cascade Classifiers are a form of ensemble learning systems. Such systems use a collection of a large number of simple classifiers in a cascade. This leads to accurate yet computationally efficient detection systems.

### What are Haar feature-based Cascade Classifiers?

Haar features are named after Haar wavelets in mathematics. The are patterns in the pixel values of an image such as edges, lines and neighbors that are used with a windowing technique to extract features from an image. Since the features could be different, a collection of specialized but simple pattern classifiers are used in a cascade to perform the feature detection.

## Sample Code

First we do the usual initializations ...

```python
# import libraries
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import time
```

# Load Cascade Files

```python
# Create the haar cascade
#cascPath = 'haar_files/haarcascade_frontalface_alt.xml'
cascPath = 'haar_files/haarcascade_frontalface_default.xml'
faceCascade = cv2.CascadeClassifier(cascPath)
print(faceCascade)
```

```python
# Read the image and show before running face detection
img = cv2.imread('images/people.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
# method to detect faces and show 
def detect_faces_show(fpath, cascade_file_path):
    # read image
    img = cv2.imread(fpath)
    # convert to gray scale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # use faceCascade object to get face coordinates from gray scale image
    fCascade = cv2.CascadeClassifier(cascade_file_path)
    faces = fCascade.detectMultiScale(
        gray,
        scaleFactor=1.2,
        minNeighbors=5,
        minSize=(30, 30),
        flags = cv2.CASCADE_SCALE_IMAGE
    )
    # length of faces which includes x, y, width, height values for each detected face
    print("Found %d faces!" % len(faces))

    # Draw a rectangle around the faces
    for (x, y, w, h) in faces:
        # use rectangle method to draw rectangle
        cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 4)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
# run method
image_path = 'images/people.jpg'

#cascPath = 'haar_files/haarcascade_frontalface_alt.xml'
cascPath = 'haar_files/haarcascade_frontalface_default.xml'

detect_faces_show(image_path, cascPath)
```
